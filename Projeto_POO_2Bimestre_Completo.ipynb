```python


class NotaAluno:
    def __init__(self, nome, matricula, notas):
        self.nome = nome
        self.matricula = matricula
        self.notas = notas

    def calcular_media(self):
        return sum(self.notas) / len(self.notas)
    
    def exibir_dados(self):
        media = self.calcular_media()
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Notas: {', '.join(map(str, self.notas))}")
        print(f"M√©dia: {media:.2f}")
    
    def adicionar_nota(self, nota):
        self.notas.append(nota)
        print(f"Nota {nota} adicionada com sucesso!")

    def remover_nota(self, nota):
        if nota in self.notas:
            self.notas.remove(nota)
            print(f"Nota {nota} removida com sucesso!")
        else:
            print(f"Nota {nota} n√£o encontrada!")

class NotaAlunoAvancado(NotaAluno):
    def __init__(self, nome, matricula, notas, curso):
        super().__init__(nome, matricula, notas)
        self.curso = curso
    
    def exibir_curso(self):
        print(f"Curso: {self.curso}")

guilherme = NotaAluno("Guilherme Refundini", "12345", [7.5, 8.0, 9.0])
pedro = NotaAluno("Pedro Ramalho", "67890", [6.0, 7.0, 8.0])

gustavo = NotaAlunoAvancado("Gustavo Fadel", "11223", [9.5, 10.0, 8.5], "Engenharia")

print("Dados do aluno Guilherme:")
guilherme.exibir_dados()
print()

print("Dados do aluno Pedro:")
pedro.exibir_dados()
print()

print("Dados do aluno Gustavo:")
gustavo.exibir_dados()
gustavo.exibir_curso() 

guilherme.adicionar_nota(9.0)
guilherme.exibir_dados()

pedro.remover_nota(7.0)
pedro.exibir_dados()
```

## PARTE 3 ‚Äì Encapsulamento e Abstra√ß√£o

**Tarefas:**
- Tornar pelo menos 1 atributo da classe principal privado (`__atributo`).

- Implementar um m√©todo que contenha l√≥gica interna (ex: c√°lculo, verifica√ß√£o).

**C√≥digo:**

```python

class NotaAluno:
    def __init__(self, nome, matricula, notas):
        self.__notas = notas
        self.nome = nome
        self.matricula = matricula
    
    def calcular_media(self):
        return sum(self.__notas) / len(self.__notas)
    
    def exibir_dados(self):
        media = self.calcular_media()
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Notas: {', '.join(map(str, self.__notas))}")
        print(f"M√©dia: {media:.2f}")

    def adicionar_nota(self, nota):
        self.__notas.append(nota)
        print(f"Nota {nota} adicionada com sucesso!")
    
    def remover_nota(self, nota):
        if nota in self.__notas:
            self.__notas.remove(nota)
            print(f"Nota {nota} removida com sucesso!")
        else:
            print(f"Nota {nota} n√£o encontrada!")
    
    def get_notas(self):
        return self.__notas
    
    def set_notas(self, novas_notas):
        if isinstance(novas_notas, list) and all(isinstance(nota, (int, float)) for nota in novas_notas):
            self.__notas = novas_notas
            print("Notas atualizadas com sucesso!")
        else:
            print("Erro: As notas precisam ser uma lista de n√∫meros v√°lidos.")

    def verificar_aprovacao(self):
        media = self.calcular_media()
        if media >= 7.0:
            return "Aprovado"
        else:
            return "Reprovado"

class NotaAlunoAvancado(NotaAluno):
    def __init__(self, nome, matricula, notas, curso):
        super().__init__(nome, matricula, notas)
        self.curso = curso
    
    def exibir_curso(self):
        print(f"Curso: {self.curso}")

guilherme = NotaAluno("Guilherme Refundini", "12345", [7.5, 8.0, 9.0])
pedro = NotaAluno("Pedro Ramalho", "67890", [6.0, 7.0, 8.0])
gustavo = NotaAlunoAvancado("Gustavo Fadel", "11223", [9.5, 10.0, 8.5], "Engenharia")

print("Dados do aluno Guilherme:")
guilherme.exibir_dados()
print(f"Aprova√ß√£o: {guilherme.verificar_aprovacao()}")
print()

print("Dados do aluno Pedro:")
pedro.exibir_dados()
print(f"Aprova√ß√£o: {pedro.verificar_aprovacao()}")
print()

print("Dados do aluno Gustavo:")
gustavo.exibir_dados()
gustavo.exibir_curso()  
print(f"Aprova√ß√£o: {gustavo.verificar_aprovacao()}")

guilherme.adicionar_nota(9.0)
guilherme.exibir_dados()

pedro.remover_nota(7.0)
pedro.exibir_dados()

print("\nUsando os m√©todos get e set para as notas de Guilherme:")
print("Notas atuais:", guilherme.get_notas())
guilherme.set_notas([8.5, 9.0, 7.5])  
guilherme.exibir_dados()
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

class NotaAluno:
    def __init__(self, nome, matricula, notas):
        self.__notas = notas
        self.nome = nome
        self.matricula = matricula
    
    def calcular_media(self):
        return sum(self.__notas) / len(self.__notas)
    
    def exibir_dados(self):
        media = self.calcular_media()
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Notas: {', '.join(map(str, self.__notas))}")
        print(f"M√©dia: {media:.2f}")
    
    def adicionar_nota(self, nota):
        self.__notas.append(nota)
        print(f"Nota {nota} adicionada com sucesso!")
    
    def remover_nota(self, nota):
        if nota in self.__notas:
            self.__notas.remove(nota)
            print(f"Nota {nota} removida com sucesso!")
        else:
            print(f"Nota {nota} n√£o encontrada!")
    
    def get_notas(self):
        return self.__notas
    
    def set_notas(self, novas_notas):
        if isinstance(novas_notas, list) and all(isinstance(nota, (int, float)) for nota in novas_notas):
            self.__notas = novas_notas
            print("Notas atualizadas com sucesso!")
        else:
            print("Erro: As notas precisam ser uma lista de n√∫meros v√°lidos.")

    def verificar_aprovacao(self):
        media = self.calcular_media()
        if media >= 7.0:
            return "Aprovado"
        else:
            return "Reprovado"
    
    def exibir_dados_completo(self):
        return f"Nome: {self.nome}, Matr√≠cula: {self.matricula}, M√©dia: {self.calcular_media():.2f}"

class NotaAlunoAvancado(NotaAluno):
    def __init__(self, nome, matricula, notas, curso):
        super().__init__(nome, matricula, notas)
        self.curso = curso
    
    def exibir_dados_completo(self):
        return f"Nome: {self.nome}, Matr√≠cula: {self.matricula}, Curso: {self.curso}, M√©dia: {self.calcular_media():.2f}"


def conectar_bd():
    return sqlite3.connect('alunos.db')


def criar_tabela():
    conn = conectar_bd()
    cursor = conn.cursor()
    cursor.execute("""
        CREATE TABLE IF NOT EXISTS alunos (
            matricula TEXT PRIMARY KEY,
            nome TEXT,
            notas TEXT,
            curso TEXT
        )
    """)
    conn.commit()
    conn.close()


def inserir_aluno(aluno):
    conn = conectar_bd()
    cursor = conn.cursor()
    
    notas_str = ','.join(map(str, aluno.get_notas())) 
    
    cursor.execute("""
        INSERT INTO alunos (matricula, nome, notas, curso) 
        VALUES (?, ?, ?, ?)
    """, (aluno.matricula, aluno.nome, notas_str, getattr(aluno, 'curso', None)))
    
    conn.commit()
    conn.close()
    print(f"Aluno {aluno.nome} inserido com sucesso no banco de dados!")


def consultar_aluno(matricula):
    conn = conectar_bd()
    cursor = conn.cursor()
    
    cursor.execute("SELECT * FROM alunos WHERE matricula=?", (matricula,))
    aluno = cursor.fetchone()
    
    if aluno:
        print(f"Aluno encontrado: Nome: {aluno[1]}, Matr√≠cula: {aluno[0]}, Notas: {aluno[2]}, Curso: {aluno[3]}")
    else:
        print(f"Aluno com matr√≠cula {matricula} n√£o encontrado.")
    
    conn.close()


def excluir_aluno(matricula):
    conn = conectar_bd()
    cursor = conn.cursor()
    
    cursor.execute("DELETE FROM alunos WHERE matricula=?", (matricula,))
    conn.commit()
    
    if cursor.rowcount > 0:
        print(f"Aluno com matr√≠cula {matricula} exclu√≠do com sucesso!")
    else:
        print(f"Aluno com matr√≠cula {matricula} n√£o encontrado.")
    
    conn.close()


guilherme = NotaAluno("Guilherme Refundini", "12345", [7.5, 8.0, 9.0])
pedro = NotaAluno("Pedro Ramalho", "67890", [6.0, 7.0, 8.0])

gustavo = NotaAlunoAvancado("Gustavo Fadel", "11223", [9.5, 10.0, 8.5], "Engenharia")

print(guilherme.exibir_dados_completo())
print(pedro.exibir_dados_completo())
print(gustavo.exibir_dados_completo())


criar_tabela()
inserir_aluno(guilherme)
inserir_aluno(pedro)
inserir_aluno(gustavo)
consultar_aluno("12345")
consultar_aluno("67890")
excluir_aluno("11223")
consultar_aluno("11223") 
```
