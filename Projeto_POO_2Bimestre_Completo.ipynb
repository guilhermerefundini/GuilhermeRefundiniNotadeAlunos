## PARTE 1 ‚Äì Classe Principal e Objetos

**Tarefas:**
- Definir e criar a classe principal com pelo menos 3 atributos.

- Criar pelo menos 2 objetos da classe principal.
- Implementar um m√©todo de exibi√ß√£o de dados.

**C√≥digo:**

```python
# Defini√ß√£o da classe principal
class Aluno:
    def __init__(self, nome, matricula, notas):
        self.nome = nome                  # Nome do aluno
        self.matricula = matricula        # N√∫mero de matr√≠cula
        self.notas = notas                # Lista de notas (ex: [8.0, 7.5, 9.0])

    # M√©todo para exibir os dados do aluno
    def exibir_dados(self):
        media = sum(self.notas) / len(self.notas)
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Notas: {self.notas}")
        print(f"M√©dia: {media:.2f}")
        print("-" * 30)

# Cria√ß√£o de objetos da classe Aluno
aluno1 = Aluno("Guilherme Refundini", "2025001", [8.5, 7.0, 9.2])
aluno2 = Aluno("Pedro Ramalho", "2025002", [6.0, 7.5, 8.0])
aluno3 = Aluno("Gustavo Fadel", "2025003", [9.0, 9.5, 10.0])

# Exibi√ß√£o dos dados dos alunos
aluno1.exibir_dados()
aluno2.exibir_dados()
aluno3.exibir_dados()
```

## PARTE 2 ‚Äì M√©todos e Heran√ßa

**Tarefas:**
- Criar pelo menos 2 m√©todos de a√ß√£o para a classe principal.
- Criar uma subclasse herdando da classe principal.
- Adicionar pelo menos 1 novo atributo e 1 novo m√©todo exclusivo na subclasse.
- Utilizar `super()` no construtor da subclasse.

**C√≥digo:**

```python
class Aluno:
    def __init__(self, nome, matricula, notas):
        self.nome = nome
        self.matricula = matricula
        self.notas = notas

    def calcular_media(self):
        return sum(self.notas) / len(self.notas)

    def exibir_dados(self):
        media = self.calcular_media()
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.matricula}")
        print(f"Notas: {self.notas}")
        print(f"M√©dia: {media:.2f}")
        print("-" * 30)

class AlunoGraduacao(Aluno):
    def __init__(self, nome, matricula, notas, curso):
        super().__init__(nome, matricula, notas)
        self.curso = curso

    def verificar_aprovacao(self):
        media = self.calcular_media()
        if media >= 7:
            return "Aprovado"
        else:
            return "Reprovado"

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Curso: {self.curso}")
        print(f"Situa√ß√£o: {self.verificar_aprovacao()}")
        print("=" * 30)

aluno1 = AlunoGraduacao("Guilherme Refundini", "2025001", [8.5, 7.0, 9.2], "Engenharia")
aluno2 = AlunoGraduacao("Pedro Ramalho", "2025002", [6.0, 7.5, 8.0], "Administra√ß√£o")
aluno3 = AlunoGraduacao("Gustavo Fadel", "2025003", [9.0, 9.5, 10.0], "Ci√™ncia da Computa√ß√£o")

aluno1.exibir_dados()
aluno2.exibir_dados()
aluno3.exibir_dados()
```

## PARTE 4 ‚Äì Polimorfismo e Banco de Dados

**Tarefas:**
- Implementar polimorfismo: sobrescrever um m√©todo na subclasse.
- Integrar o sistema com banco de dados (MySQL ou SQLite).
- Implementar inser√ß√£o (INSERT), consulta (SELECT) e exclus√£o (DELETE) de registros no banco de dados.

**C√≥digo:**

```python
import sqlite3

class Aluno:
    def __init__(self, nome, matricula, notas):
        self.nome = nome
        self.__matricula = matricula
        self.notas = notas

    def get_matricula(self):
        return self.__matricula

    def set_matricula(self, nova_matricula):
        if isinstance(nova_matricula, str) and nova_matricula.isdigit():
            self.__matricula = nova_matricula
        else:
            print("Matr√≠cula inv√°lida. Deve conter apenas n√∫meros.")

    def calcular_media(self):
        return sum(self.notas) / len(self.notas)

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.get_matricula()}")
        print(f"Notas: {self.notas}")
        print(f"M√©dia: {self.calcular_media():.2f}")
        print("-" * 30)

class AlunoGraduacao(Aluno):
    def __init__(self, nome, matricula, notas, curso):
        super().__init__(nome, matricula, notas)
        self.curso = curso

    def verificar_aprovacao(self):
        return "Aprovado" if self.calcular_media() >= 7 else "Reprovado"

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Curso: {self.curso}")
        print(f"Situa√ß√£o: {self.verificar_aprovacao()}")
        print("=" * 30)

# --- Banco de Dados ---

# Cria√ß√£o/Conex√£o com o banco
conn = sqlite3.connect("alunos.db")
cursor = conn.cursor()

# Cria√ß√£o da tabela (se n√£o existir)
cursor.execute("""
CREATE TABLE IF NOT EXISTS alunos (
    nome TEXT,
    matricula TEXT PRIMARY KEY,
    notas TEXT,
    curso TEXT
)
""")
conn.commit()

# Fun√ß√µes para manipula√ß√£o de dados
def inserir_aluno(aluno):
    notas_str = ",".join(map(str, aluno.notas))
    cursor.execute("INSERT INTO alunos (nome, matricula, notas, curso) VALUES (?, ?, ?, ?)",
                   (aluno.nome, aluno.get_matricula(), notas_str, aluno.curso))
    conn.commit()

def listar_alunos():
    cursor.execute("SELECT * FROM alunos")
    alunos = cursor.fetchall()
    for a in alunos:
        nome, matricula, notas_str, curso = a
        notas = list(map(float, notas_str.split(",")))
        aluno = AlunoGraduacao(nome, matricula, notas, curso)
        aluno.exibir_dados()

def excluir_aluno(matricula):
    cursor.execute("DELETE FROM alunos WHERE matricula = ?", (matricula,))
    conn.commit()

# --- Exemplo de uso ---

aluno1 = AlunoGraduacao("Guilherme Refundini", "2025001", [8.5, 7.0, 9.2], "Engenharia")
aluno2 = AlunoGraduacao("Pedro Ramalho", "2025002", [6.0, 7.5, 8.0], "Administra√ß√£o")
aluno3 = AlunoGraduacao("Gustavo Fadel", "2025003", [9.0, 9.5, 10.0], "Ci√™ncia da Computa√ß√£o")

inserir_aluno(aluno1)
inserir_aluno(aluno2)
inserir_aluno(aluno3)

print(">>> ALUNOS NO BANCO DE DADOS:")
listar_alunos()

excluir_aluno("2025002")

print("\n>>> AP√ìS EXCLUS√ÉO DO PEDRO RAMALHO:")
listar_alunos()

conn.close()
```

## PARTE 5 ‚Äì Projeto Final e Organiza√ß√£o

```python
import sqlite3

# -------------------------
# CLASSE PRINCIPAL Aluno
# -------------------------
class Aluno:
    def __init__(self, nome, matricula, notas):
        self.nome = nome
        self.__matricula = matricula
        self.notas = notas

    def get_matricula(self):
        return self.__matricula

    def set_matricula(self, nova_matricula):
        if isinstance(nova_matricula, str) and nova_matricula.isdigit():
            self.__matricula = nova_matricula
        else:
            print("Matr√≠cula inv√°lida!")

    def calcular_media(self):
        return sum(self.notas) / len(self.notas)

    def exibir_dados(self):
        print(f"Nome: {self.nome}")
        print(f"Matr√≠cula: {self.get_matricula()}")
        print(f"Notas: {self.notas}")
        print(f"M√©dia: {self.calcular_media():.2f}")

# -------------------------
# SUBCLASSE AlunoGraduacao
# -------------------------
class AlunoGraduacao(Aluno):
    def __init__(self, nome, matricula, notas, curso):
        super().__init__(nome, matricula, notas)
        self.curso = curso

    def verificar_aprovacao(self):
        return "Aprovado" if self.calcular_media() >= 7 else "Reprovado"

    def exibir_dados(self):
        super().exibir_dados()
        print(f"Curso: {self.curso}")
        print(f"Situa√ß√£o: {self.verificar_aprovacao()}")

# -------------------------
# BANCO NA MEM√ìRIA
# -------------------------
conn = sqlite3.connect(":memory:")
cursor = conn.cursor()

cursor.execute("""
CREATE TABLE alunos (
    nome TEXT,
    matricula TEXT PRIMARY KEY,
    notas TEXT,
    curso TEXT
)
""")
conn.commit()

# -------------------------
# FUN√á√ïES BANCO
# -------------------------
def inserir_aluno(aluno):
    notas_str = ",".join(map(str, aluno.notas))
    try:
        cursor.execute("INSERT INTO alunos (nome, matricula, notas, curso) VALUES (?, ?, ?, ?)",
                       (aluno.nome, aluno.get_matricula(), notas_str, aluno.curso))
        conn.commit()
    except sqlite3.IntegrityError:
        print("Erro: Matr√≠cula j√° cadastrada!")

def listar_alunos():
    cursor.execute("SELECT * FROM alunos")
    registros = cursor.fetchall()
    if not registros:
        print("Nenhum aluno cadastrado.")
        return
    for nome, matricula, notas_str, curso in registros:
        notas = list(map(float, notas_str.split(",")))
        aluno = AlunoGraduacao(nome, matricula, notas, curso)
        aluno.exibir_dados()
        print("-" * 30)

def excluir_aluno(matricula):
    cursor.execute("DELETE FROM alunos WHERE matricula = ?", (matricula,))
    conn.commit()

# -------------------------
# INTERFACE SIMPLES
# -------------------------
def menu():
    while True:
        print("\n=== MENU ALUNOS ===")
        print("1 - Inserir aluno")
        print("2 - Listar alunos")
        print("3 - Excluir aluno por matr√≠cula")
        print("0 - Sair")
        escolha = input("Op√ß√£o: ")

        if escolha == "1":
            nome = input("Nome: ")
            matricula = input("Matr√≠cula (somente n√∫meros): ")
            notas_input = input("Notas separadas por v√≠rgula (ex: 7.5,8.0,9.0): ")
            try:
                notas = [float(n) for n in notas_input.split(",")]
            except:
                print("Notas inv√°lidas! Tente novamente.")
                continue
            curso = input("Curso: ")
            aluno = AlunoGraduacao(nome, matricula, notas, curso)
            inserir_aluno(aluno)
        elif escolha == "2":
            listar_alunos()
        elif escolha == "3":
            mat = input("Digite a matr√≠cula para exclus√£o: ")
            excluir_aluno(mat)
            print("Aluno exclu√≠do, se matr√≠cula existia.")
        elif escolha == "0":
            print("Saindo...")
            break
        else:
            print("Op√ß√£o inv√°lida! Tente novamente.")

# -------------------------
# EXECU√á√ÉO
# -------------------------
if __name__ == "__main__":
    menu()
    conn.close()
```
